```
def draw_pattern(number):
    character= '*'
    for i in range(1,number,1):
        print(character * i )
    for i in range(number,0,-1):
        print(character * i )

draw_pattern(5)
```

```
input_string='Vikas Mahajan'

def reverse_string_loop(s):
    s1 = ''
    for c in s:
        s1 = c + s1  
    return s1

reverse_string_loop(input_string)
```
