```
#Write a Python Program(with class concepts) to find the area of the triangle using the below formula.
#area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
#Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.
import math
class Triangle:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    
    def calTrianleParameter(self):
        return (self.a+self.b+self.c)/2

class EquilateralTriangle(Triangle):
    def __init__(self,a):
        b=a
        c=a
        super().__init__(a,b,c)

    def area(self):
        s=super().calTrianleParameter()
        return math.sqrt(s*(s-self.a)*(s-self.b)*(s-self.c)) 

equi=EquilateralTriangle(10)
equi.area()
```

```
#Write a function filter_long_words() that takes a list of words and an integer n and returns
#the list of words that are longer than n.

l =['asd','dd','asdfg','xcvbn','vikas','Ajay']
def filter_long_words(lst):
    return [item for item in lst if len(item)>4]

print(filter_long_words(l))

list(filter(lambda a: len(a)>3, l))
```

```
#Write a Python program using function concept that maps list of words into a list of integers
#representing the lengths of the corresponding words.
l =['asd','dd','asdfg','xcvbn','vikas','Ajay']
def Calculate_len_str(lst):    
    return [item+" : " +str(len(item)) for item in lst]

def Calculate_len_str1(lst):
    return [(item,len(item)) for item in lst]

print(Calculate_len_str(l))
Calculate_len_str1(l)
```

```
#Write a Python function which takes a character (i.e. a string of length 1) and returns True if
#it is a vowel, False otherwise.

vowels=['a','e','i','o','u']
def IsVowel(char):
    return True if char in vowels else False
    
print(IsVowel('a'))
print(IsVowel('z'))
```
